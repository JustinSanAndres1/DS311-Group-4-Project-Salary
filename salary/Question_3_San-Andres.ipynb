```python
# Imported necessary libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas.plotting as pdplt

# Load the salary data.
salary_data_load = 'salary_data_states.xlsx'

salary_data = pd.read_excel(salary_data_load)

# Load data that contains the cost index.
cost_index_load = 'data.csv'

cost_index_data = pd.read_csv(cost_index_load)

# Merged the two data sets to make it easier to work with. 
salary_cost_index = pd.merge(salary_data, cost_index_data, left_on='WORK_STATE', right_on='state')

salary_cost_index
```

```python
# 3a. 
# To start, we need to be able to create columns where we can see where the instances happen in order by date.
# Created a variable that formats date data into the months, days, and years. 
date_format = '%m/%d/%Y'

# Created bins using date_format for use in question(s).
bins = pd.to_datetime(['01/01/2010', '01/01/2012', '01/01/2014', '01/01/2016'], format=date_format)

labels = ['01/01/2010 - 12/31/2011', '01/01/2012 - 12/31/2013', '01/01/2014 - 12/31/2015']

# Created new column case_count, where CASE_RECEIVED_DATE is formated into our bins. 
salary_cost_index['case_count'] = pd.cut(pd.to_datetime(salary_cost_index['CASE_RECEIVED_DATE'], format=date_format), bins, labels=labels, include_lowest=True)

# To answer this question, I decided to create a pivot table using PAID_WAGE_PER_YEAR (Assumed this was what the company offered.) 
# and PREVAILING_WAGE_SUBMITTED, it was then indexed by JOB_TITLE_SUBGROUP and created a column for our new column that carries date ranges. 
offer_difference = pd.pivot_table(salary_cost_index, values=['PAID_WAGE_PER_YEAR', 'PREVAILING_WAGE_SUBMITTED'], index='JOB_TITLE_SUBGROUP', columns='case_count', aggfunc='mean').round(2)

offer_difference
```

Here we see that the prevailing wage submitted is considerably lower than the paid wage per year, seeing drastic changes when we observe assistant professors, management consultants, and teachers. Categories with the lowest change seemingly being data scientists and software engineers. 

```python
# 3b. 
# Approached by declaring a new column ('is_over_paid') in our data set that checks if the submitted prevailing wage is more than or equal to the paid wage per year. 
# Made into a bool so that our data efficiently tells observers instances where an employee for specific subgroups were or were not over-paid.  
salary_cost_index['is_over_paid'] = (salary_cost_index['PREVAILING_WAGE_SUBMITTED'] >= salary_cost_index['PAID_WAGE_PER_YEAR']).astype(bool)

over_under = salary_cost_index.groupby(['JOB_TITLE_SUBGROUP', 'is_over_paid']).size()

over_under
```

Though in all sub-categories we can see the majority of people were not over paid, a staggering portion of business analysts, software engineers, and teachers were reported to be over-paid. 

```python
# 3c. 
# Top 5 Companies that overpay
# Using the new column is_over_paid, created a variable that groups the data frame by EMPLOYER_NAME and checks the number of times a company has over-paid. 
company_over = salary_cost_index[salary_cost_index['is_over_paid'] == True].groupby('EMPLOYER_NAME').size().nlargest(5)

company_over
```

```python
# Top 5 Companies that underpay
# Using the new column is_over_paid, created a variable that groups the data frame by EMPLOYER_NAME and checks the number of times a company has under-paid. 
company_under = salary_cost_index[salary_cost_index['is_over_paid'] == False].groupby('EMPLOYER_NAME').size().nlargest(5)

company_under
```

```python
# 3d. 
# Created variables that take either PAID_WAGE_PER_YEAR or PREVAILING_WAGE_SUBMITTED, and calculates the wage after considering costIndex
salary_cost_index['WAGE_COST_LIVING_SUBMITTED_CALC'] = (salary_cost_index['PAID_WAGE_PER_YEAR'] / salary_cost_index['costIndex']) * 100

salary_cost_index['PREVAILING_WAGE_COST_LIVING_SUBMITTED_CALC'] = (salary_cost_index['PREVAILING_WAGE_SUBMITTED'] / salary_cost_index['costIndex']) * 100

# Using the same method to make a pivot table for 3a, displays resutlts.
offer_difference = pd.pivot_table(salary_cost_index, values=['WAGE_COST_LIVING_SUBMITTED_CALC', 'PREVAILING_WAGE_COST_LIVING_SUBMITTED_CALC'], index='JOB_TITLE_SUBGROUP', columns='case_count', aggfunc='mean').round(2)

offer_difference
```
